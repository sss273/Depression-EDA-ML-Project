```python
X = df.drop(['Participant_ID','PHQ8_Score', 'PHQ8_Binary', 'Gender'], axis='columns')
```

```python
y = df.PHQ8_Binary
```

```python
from sklearn.preprocessing import LabelEncoder
# label encoding
```

```python
le_PHQ8_NoInterest = LabelEncoder()
le_PHQ8_Depressed = LabelEncoder()
le_PHQ8_Sleep = LabelEncoder()
le_PHQ8_Tired = LabelEncoder()
le_PHQ8_Appetite = LabelEncoder()
le_PHQ8_Failure = LabelEncoder()
le_PHQ8_Concentrating = LabelEncoder()
le_PHQ8_Moving = LabelEncoder()
```

```python
X['PHQ8_NoInterest_n'] = le_PHQ8_NoInterest.fit_transform(X['PHQ8_NoInterest'])
X['PHQ8_Depressed_n'] = le_PHQ8_Depressed.fit_transform(X['PHQ8_Depressed'])
X['PHQ8_Sleep_n'] = le_PHQ8_Sleep.fit_transform(X['PHQ8_Sleep'])
X['PHQ8_Tired_n'] = le_PHQ8_Tired.fit_transform(X['PHQ8_Tired'])
X['PHQ8_Appetite_n'] = le_PHQ8_Appetite.fit_transform(X['PHQ8_Appetite'])
X['PHQ8_Failure_n'] = le_PHQ8_Failure.fit_transform(X['PHQ8_Failure'])
X['PHQ8_Concentrating_n'] = le_PHQ8_Concentrating.fit_transform(X['PHQ8_Concentrating'])
X['PHQ8_Moving_n'] = le_PHQ8_Moving.fit_transform(X['PHQ8_Moving'])
X.head()
```

```python
X_n = X.drop(['PHQ8_NoInterest', 'PHQ8_Depressed', 'PHQ8_Sleep', 'PHQ8_Tired', 'PHQ8_Appetite', 'PHQ8_Failure', 'PHQ8_Concentrating', 'PHQ8_Moving'], axis = 'columns')
X_n
```

```python
y.tail(100)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_n, y, test_size=0.30, random_state=101)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_n, y)
```

```python
model.score(X_n, y)
```
